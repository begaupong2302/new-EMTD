```python
count = 0

for dirname, _, filenames in os.walk('../../abc'):
    for filename in filenames:
        data = pd.read_csv(os.path.join(dirname, filename))
        full_dts = pd.concat([full_dts, data])
        count+=1
        print(os.path.join(dirname, filename))

print(count)
```
